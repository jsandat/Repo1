```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
df=pd.read_excel("nuevodos.xlsx")
```

```python
df
```

```python
df2=df[(df["MOTORIZADO"]=="SIN UBICAR")]
```

```python
df2
```

```python
df2=df2.groupby("FECHA").count().sort_values("FECHA", ascending=True)
df2
```

```python
df2.drop("Unnamed: 0",axis=1,inplace=True)
df2.drop("MOTORIZADO",axis=1,inplace=True)
df2.drop("PROVEEDOR",axis=1,inplace=True)
```

```python
df2
```

```python
filt=df2.loc[df2["SERIES"]>10]
```

```python
filt.to_excel("final.xlsx")
df3=pd.read_excel("final.xlsx")
```

```python
df3["FECHA"]=df3["FECHA"].astype(str)
df3
```

```python
fig, ax=plt.subplots()
bars=ax.bar(df3["FECHA"],df3["SERIES"],color="green")

ax.bar_label(bars, label_type="edge", color="black")
plt.xticks(rotation=20)
plt.title("Chips no ubicados")
```
